```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC, LinearSVC
from xgboost import XGBClassifier
from sklearn.metrics import confusion_matrix as cm

from sklearn.metrics import make_scorer, accuracy_score, roc_auc_score 
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import train_test_split
```

```python
features = ["acousticness", "danceability", "duration_ms", "energy", "instrumentalness", "key", "liveness", 
            "mode", "speechiness", "tempo", "time_signature", "valence"]
```

# Training phase

```python
training = dataframe.sample(frac = 0.8,random_state = 420)
X_train = training[features]
y_train = training['popularity']
X_test = dataframe.drop(training.index)[features]
```

```python
X_train, X_valid, y_train, y_valid = train_test_split(X_train, y_train, test_size = 0.2, random_state = 420)
```

# Confusion matrix

```python
from sklearn.metrics import confusion_matrix 
from sklearn.metrics import confusion_matrix

# train a classification model on the training set and make predictions on the validation set
model = RandomForestClassifier()
model.fit(X_train, y_train)
y_pred = model.predict(X_valid)

# calculate confusion matrix
cm = confusion_matrix(y_valid, y_pred)
print(cm)
```

```python
sns.heatmap(cm,annot=True,fmt="d") 
plt.show()
```

```python
RFC_Model = RandomForestClassifier()
RFC_Model.fit(X_train, y_train)
RFC_Predict = RFC_Model.predict(X_valid)
RFC_Accuracy = accuracy_score(y_valid, RFC_Predict)
print("Accuracy: " + str(RFC_Accuracy))

RFC_AUC = roc_auc_score(y_valid, RFC_Predict) 
print("AUC: " + str(RFC_AUC))
```

```python
KNN_Model = KNeighborsClassifier()
KNN_Model.fit(X_train, y_train)
KNN_Predict = KNN_Model.predict(X_valid)
KNN_Accuracy = accuracy_score(y_valid, KNN_Predict)
print("Accuracy: " + str(KNN_Accuracy))

KNN_AUC = roc_auc_score(y_valid, KNN_Predict) 
print("AUC: " + str(KNN_AUC))
```

```python
import tkinter as tk
from tkinter import ttk
import pandas as pd
import random
from PIL import Image, ImageTk
from tkinter import Label
from tkinter.scrolledtext import ScrolledText
from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import LabelEncoder


# Load the music dataset
try:
    music_data = pd.read_csv("SpotifyFeatures.csv")
except FileNotFoundError:
    print("Error: music dataset file not found")
    exit()
except pd.errors.ParserError:
    print("Error: invalid music dataset file format")
    exit()

# Instantiate and fit a LabelEncoder object to the 'genre' column of 'music_data'
le = LabelEncoder()
le.fit(music_data['genre'])

# Create the GUI window
root = tk.Tk()
root.title("Music Predictor")

# Set up the background image
bg_image = Image.open("th.jpg")
screen_width = root.winfo_screenwidth()
screen_height = root.winfo_screenheight()
bg_image = bg_image.resize((screen_width, screen_height), Image.ANTIALIAS)
bg_photo = ImageTk.PhotoImage(bg_image)
bg_label = Label(root, image=bg_photo)
bg_label.place(relx=0.5, rely=0.5, anchor=tk.CENTER)


# Create the style for the widgets
style = ttk.Style()
style.configure('TLabel', font=('Arial', 12), padding=10)
style.configure('TButton', font=('Arial', 12), padding=10)
style.configure('TMenubutton', font=('Arial', 12), padding=10)
style.configure('TEntry', font=('Arial', 12), padding=10)

# Create the genre label and drop-down menu
genre_label = ttk.Label(root, text="Select a genre:")
genre_label.pack(side=tk.TOP)
genres = sorted(list(set(le.transform(music_data["genre"]))))
genre_var = tk.StringVar(value=le.inverse_transform([genres[0]])[0])
genre_menu = ttk.OptionMenu(root, genre_var, *[le.inverse_transform([i])[0] for i in genres])
genre_menu.pack(side=tk.TOP)

# Define the function to predict music based on user input
def predict_music():
    # Get the selected genre
    genre = genre_var.get()

    # Filter the music dataset to get songs in the selected genre
    genre_songs = music_data[music_data["genre"] == genre]

    # Check if any songs were found in the selected genre
    if len(genre_songs) == 0:
        result_label.config(text="No songs found in selected genre")
        return

    # Get the number of songs to recommend
    num_songs = int(num_songs_entry.get())

    # Select a number of random songs from the filtered dataset
    if len(genre_songs) < num_songs:
        result_label.config(text=f"Not enough songs in selected genre (found {len(genre_songs)} song(s))")
        return
    songs = random.sample(list(genre_songs["track_name"]), num_songs)

    # Prepare the data for training
    genre_data = music_data[music_data["genre"] == genre]
    X = genre_data.drop(columns=["genre", "artist_name", "track_name", "track_id"])
    y = genre_data["genre"]
    enc = OneHotEncoder()
    X_enc = enc.fit_transform(X)

    # Train the model
    knn = KNeighborsClassifier(n_neighbors=5)
    knn.fit(X_enc, y)

    # Predict the songs
    song_indexes = []
    for song in songs:
        index = list(genre_songs["track_name"]).index(song)
        song_indexes.append(index)
    X_pred = X_enc[song_indexes]
    y_pred = knn.predict(X_pred)

    # Display the result to the user
    result_songs = []
    for song, prediction in zip(songs, y_pred):
        result_songs.append(f"{song} ({prediction})")
    song_list = "\n- ".join(result_songs)
    result_label.configure(state='normal')
    result_label.delete('1.0', tk.END)
    result_label.insert(tk.END, f"Recommended songs:\n- {song_list}")
    result_label.configure(state='disabled')

#Create a label for the number of songs entry field
num_songs_label = ttk.Label(root, text="Enter the number of songs to recommend:")
num_songs_label.pack(side=tk.TOP)

#Create an entry field for the number of songs
num_songs_var = tk.StringVar(value="5")
num_songs_entry = ttk.Entry(root, textvariable=num_songs_var)
num_songs_entry.pack(side=tk.TOP)

#Create a button to predict music based on user input
predict_button = ttk.Button(root, text="Predict", command=predict_music)
predict_button.pack(side=tk.TOP, pady=10)

# Customize the result label
result_label = ScrolledText(root, font=('Arial', 12), wrap=tk.WORD)
result_label.place(relx=0.5, rely=0.5, anchor=tk.CENTER, width=500, height=200)


#Start the GUI loop
root.mainloop()
```
